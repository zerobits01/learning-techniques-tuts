```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.colors import ListedColormap
from sklearn import neighbors
```

```python
plt.rcParams["figure.figsize"] = [7.00, 3.50]
plt.rcParams["figure.autolayout"] = True

one_neighs = 1
three_neighs=3;
X = np.array([(1,2), (2,1), (2,2), (2,3), (3,1), (3,2)])
y = [2, 2, 1, 2, 1, 2]
h = .02

cmap_light = ListedColormap(['orange', 'cyan'])
cmap_bold = ['darkorange', 'c'] # , 'darkblue'


clf1 = neighbors.KNeighborsClassifier(one_neighs, weights='uniform')
clf1.fit(X, y)

clf3 = neighbors.KNeighborsClassifier(three_neighs, weights='uniform')
clf3.fit(X, y)


x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, h),
np.arange(y_min, y_max, h))
Z1 = clf1.predict(np.c_[xx.ravel(), yy.ravel()])
Z1 = Z1.reshape(xx.shape)

plt.figure()

plt.contourf(xx, yy, Z1, cmap=cmap_light)

sns.scatterplot(x=X[:, 0], y=X[:, 1],
palette=cmap_bold, alpha=1.0, edgecolor="black")

plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())

plt.title("2-Class classification (k = %i, 'uniform' = '%s')"
% (3, 'uniform'))

plt.show()


Z3 = clf3.predict(np.c_[xx.ravel(), yy.ravel()])
Z3 = Z3.reshape(xx.shape)

plt.figure()

plt.contourf(xx, yy, Z3, cmap=cmap_light)

sns.scatterplot(x=X[:, 0], y=X[:, 1],
palette=cmap_bold, alpha=1.0, edgecolor="black")

plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())

plt.title("2-Class classification (k = %i, 'uniform' = '%s')"
% (3, 'uniform'))

plt.show()
```
