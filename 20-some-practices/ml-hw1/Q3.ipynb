# 3.a)

```python
from sklearn.datasets import make_moons
import numpy as np
```

# 3.b)

```python

# Generate the two moons dataset
data, _ = make_moons(n_samples=10000, noise=0.05)

# Pick 500 random samples
random_samples = data[np.random.choice(data.shape[0], 500, replace=False)]
```

# 3.c)

To find the two clusters of data using hierarchical clustering in Python, you can use the AgglomerativeClustering function from the scikit-learn library


```python
from sklearn.cluster import AgglomerativeClustering
import matplotlib.pyplot as plt

# Set the number of clusters
number_of_clusters = 2

clustering = AgglomerativeClustering(n_clusters=number_of_clusters).fit(random_samples)
```

# 3.d)

```python
# Visualize the clustering
plt.scatter(random_samples[:,0], random_samples[:,1], c=clustering.labels_, cmap='viridis')
plt.show()
```

# 3.e)

```python
m = 5  # Number of points to select from each cluster
representative_points = []
# representative_points_1 = []
# rndm_indices = np.random.randint(0, len(random_samples), m)
for cluster in range(2):  # Iterate over each cluster
    cluster_points = random_samples[clustering.labels_ == cluster]  # Select points belonging to the cluster
    selected_points = cluster_points[np.random.choice(cluster_points.shape[0], m, replace=False)]  # Randomly select m points
    print(cluster)
    print(selected_points)
    representative_points.append(selected_points)

# Plot the representation points
representative_points = np.concatenate(representative_points)  # Combine the representation points from both clusters
plt.scatter(random_samples[:, 0], random_samples[:, 1], c=clustering.labels_, cmap='viridis')  # Plot all data points with cluster colors
plt.scatter(representative_points[:, 0], representative_points[:, 1], c='red', marker='x', label='Representation Points')  # Plot representation points
plt.legend()
plt.show()
```

# 3.f)

```python
cluster_labels = clustering.labels_

# Calculate the center of each cluster based on the representation points
cluster_centers = np.array([random_samples[cluster_labels == i].mean(axis=0) for i in range(2)])

for i, item in zip(range(len(representative_points)), representative_points):
    if i < 5:
        cluster_id = 0
    else:
        cluster_id = 1
    center = cluster_centers[cluster_id]
    representative_points[i] = representative_points[i] + 0.1 * (center - representative_points[i])

# Plot the updated representation points
plt.figure(figsize=(10, 7))
plt.scatter(random_samples[:, 0], random_samples[:, 1], c=clustering.labels_, cmap='viridis', label='All Points')
# plt.scatter(data[:, 0], data[:, 1], c=clustering.labels_, cmap='viridis', label='All Points')
plt.scatter(representative_points[:, 0], representative_points[:, 1], c='red', label='Updated Representation Points')
plt.legend()
plt.show()
```

# 3.g)

```python
distances = np.sqrt(((data[:, np.newaxis] - representative_points) ** 2).sum(axis=2))

# Find the index of the nearest representation point for each data point
nearest_representation_indices = np.argmin(distances, axis=1)
```

```python
new_labels = [0 if index < 5 else 1 for index in nearest_representation_indices]
# new_labels
```
