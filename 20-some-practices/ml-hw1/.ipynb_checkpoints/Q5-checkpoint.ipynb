```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import cv2

# Function to compress the image using k-means clustering
def compress_image(input_image, output_image, clusters):
    # Read the image
    image = cv2.imread(input_image)
    image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Reshape the image to be a list of pixels
    pixels = image.reshape(-1, 3)

    # Create the k-means object
    kmeans = KMeans(n_clusters=clusters, random_state=0)

    # Fit the k-means object to the data
    kmeans.fit(pixels)

    # Get the cluster centroids and labels
    new_colors = kmeans.cluster_centers_
    labels = kmeans.labels_

    # Create the quantized image
    quantized_image = new_colors[labels].reshape(image.shape)

    # Save the compressed image
    plt.imsave(output_image, quantized_image.astype(np.uint8))
```

```python
image = cv2.imread('./Q_5/tiger.png')
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
plt.imshow(image)
```

```python
compress_image('./Q_5/tiger.png', './Q_5/compressed_tiger.png', 18)
```
